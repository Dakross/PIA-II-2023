```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv', usecols=['Concepto_asociado','Red_Social'])
df['Red_Social'].unique()
```

```python
from hola import matriz_peso
```

```python
Threads = df[df['Red_Social'] == 'Threads']
Threads.rename(columns={'Concepto_asociado':'Polaridades'}, inplace=True)

Threads.head(10)
```

```python
valores = Threads['Polaridades'].value_counts()
valores
```

```python
matriz_peso.info()
```

```python
Threads.info()
```

```python
matriz_peso['Apariciones'] = matriz_peso['Polaridades'].map(valores)
matriz_peso.head(281)
```

```python
arreglo_polaridades_nulas = matriz_peso.loc[matriz_peso['Apariciones'].isnull(), 'Polaridades'].to_numpy()
print(arreglo_polaridades_nulas)
```

```python
for polaridad in arreglo_polaridades_nulas:
    valor_a_buscar = polaridad

    if valor_a_buscar in valores:
        print(f"{valor_a_buscar} existe en valores.")
    else:
        print(f"{valor_a_buscar} no existe en valores.")

```

```python
matriz_peso = matriz_peso.dropna(subset = 'Apariciones')
matriz_peso.head()
```
